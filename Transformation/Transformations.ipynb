```python
import numpy as np
import cv2
import matplotlib.pyplot as plt
```

```python
# Read the image
img = cv2.imread('jungkook.jpg')
assert img is not None, "File could not be read, check with os.path.exists()"

# Resize using scaling factors
res1 = cv2.resize(img, None, fx=2, fy=2, interpolation=cv2.INTER_CUBIC)

# Alternatively, resize using calculated dimensions
height, width = img.shape[:2]
res2 = cv2.resize(img, (2 * width, 2 * height), interpolation=cv2.INTER_CUBIC)

# Display the original and resized images
cv2.imshow('Original Image', img)
cv2.imshow('Resized Image (fx=2, fy=2)', res1)
cv2.imshow('Resized Image (calculated dimensions)', res2)

# Wait until a key is pressed and then close the windows
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
# Translation

# Read the image in grayscale
img = cv2.imread('jungkook.jpg', cv2.IMREAD_GRAYSCALE)


# Get image dimensions
rows, cols = img.shape

# Define the translation matrix
M = np.float32([[1, 0, 100], [0, 1, 50]])

# Apply the affine transformation
dst = cv2.warpAffine(img, M, (cols, rows))

# Display the original and translated images using Matplotlib

plt.subplot(1, 2, 1)
plt.imshow(img, cmap='gray')
plt.title('Original Image')
plt.axis('off')  # Hide axis

plt.subplot(1, 2, 2)
plt.imshow(dst, cmap='gray')
plt.title('Translated Image')
plt.axis('off')  # Hide axis

plt.tight_layout()
plt.show()
```

```python
# Rotation
# Read the image in grayscale
img = cv2.imread('jungkook.jpg', cv2.IMREAD_GRAYSCALE)

# Get image dimensions
rows, cols = img.shape

# Get the rotation matrix: center, angle, scale
M = cv2.getRotationMatrix2D(((cols - 1) / 2.0, (rows - 1) / 2.0), 90, 1)

# Apply the affine transformation for rotation
dst = cv2.warpAffine(img, M, (cols, rows))

# Display the original and rotated images using Matplotlib

plt.subplot(1, 2, 1)
plt.imshow(img, cmap='gray')
plt.title('Original Image')
plt.axis('off')  # Hide axis

plt.subplot(1, 2, 2)
plt.imshow(dst, cmap='gray')
plt.title('Rotated Image')
plt.axis('off')  # Hide axis

plt.tight_layout()
plt.show()
```

```python
import numpy as np
import cv2
import matplotlib.pyplot as plt
```

```python
# Load the image
img = cv2.imread('jungkook.jpg')


# Split the image into Blue, Green, and Red channels
b, g, r = cv2.split(img)
    
# Display the individual channels
plt.figure(figsize=(10, 5))

plt.subplot(1, 3, 1)
plt.imshow(b, cmap='gray')
plt.title('Blue Channel')
plt.axis('off')

plt.subplot(1, 3, 2)
plt.imshow(g, cmap='gray')
plt.title('Green Channel')
plt.axis('off')

plt.subplot(1, 3, 3)
plt.imshow(r, cmap='gray')
plt.title('Red Channel')
plt.axis('off')

plt.show()
```

```python
# Merge the channels back into a single image
merged_img = cv2.merge((b, g, r))

# Convert the merged image from BGR to RGB for displaying with matplotlib
merged_img_rgb = cv2.cvtColor(merged_img, cv2.COLOR_BGR2RGB)

# Display the merged image
plt.imshow(merged_img_rgb)
plt.title('Merged Image')
plt.axis('off')
plt.show()
```

```python
import numpy as np
# Load the image
img = cv2.imread('jungkook.jpg')


# Split the image into Blue, Green, and Red channels
b, g, r = cv2.split(img)

# Create blank images to visualize channels in color
blank = cv2.merge((b, b, b))
blue_channel = cv2.merge((b, np.zeros_like(b), np.zeros_like(b)))  # Blue color visualization
green_channel = cv2.merge((np.zeros_like(g), g, np.zeros_like(g)))  # Green color visualization
red_channel = cv2.merge((np.zeros_like(r), np.zeros_like(r), r))    # Red color visualization

# Convert to RGB for matplotlib display
blank_rgb = cv2.cvtColor(blank, cv2.COLOR_BGR2RGB)
blue_rgb = cv2.cvtColor(blue_channel, cv2.COLOR_BGR2RGB)
green_rgb = cv2.cvtColor(green_channel, cv2.COLOR_BGR2RGB)
red_rgb = cv2.cvtColor(red_channel, cv2.COLOR_BGR2RGB)

# Display each channel as grayscale and color
plt.subplot(2, 2, 1)
plt.imshow(b, cmap='gray')
plt.title('Blue Channel (Grayscale)')
plt.axis('off')

plt.subplot(2, 2, 2)
plt.imshow(blue_rgb)
plt.title('Blue Channel (Color)')
plt.axis('off')
```

```python
plt.subplot(2, 2, 3)
plt.imshow(g, cmap='gray')
plt.title('Green Channel (Grayscale)')
plt.axis('off')

plt.subplot(2, 2, 4)
plt.imshow(green_rgb)
plt.title('Green Channel (Color)')
plt.axis('off')

plt.show()
```

```python
plt.subplot(1, 2, 1)
plt.imshow(r, cmap='gray')
plt.title('Red Channel (Grayscale)')
plt.axis('off')

plt.subplot(1, 2, 2)
plt.imshow(red_rgb)
plt.title('Red Channel (Color)')
plt.axis('off')

plt.show()
```

```python
# Load an image
img = cv2.imread('jungkook.jpg')
# Set the Red channel to 0
# Setting the Red channel to 0 means that no red color will be present in the image
# effectively removing all red intensity from the image.
img[:, :, 2] = 0
    
# Convert the modified image from BGR to RGB for correct display with matplotlib
img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
    
# Display the modified image
plt.imshow(img_rgb)
plt.title('Image without Red Channel')
plt.axis('off')
plt.show()
```
